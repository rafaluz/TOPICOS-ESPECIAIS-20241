# Relembrando um pouco os conceitos de programação



“Algoritmo é uma sequência de passos que visa atingir um objetivo bem definido” (FORBELLONE, 1999)

**Características dos Algoritmos**

* Finitude: um algoritmo tem de terminar ao fim de um número finito de passos;
* Clareza/definição: cada passo do algoritmo tem de ser definido com precisão, sem ambiguidades;
* Entrada: um algoritmo pode ter zero ou mais entradas;
* Saídas: um algoritmo tem uma ou mais saídas;
* Eficácia/efetividade: todas as operações feitas por um algoritmo têm de ser suficientemente básicas.

**Programa é a codificação de um algoritmo em uma determinada linguagem de programação** (ASCENCIO, 1999)

## Paradigmas de programação

* Existem diversas linguagens de programação e cada uma possui um paradigma de programação. 
* Esse paradigma rege o modo como o programador expressará os passos (algoritmo) do processo que deseja automatizar.
* Paradigma imperativo e declarativo



# Imperativo
Criada por John von Neumann (Arquitetura de von Neumann) prega que se deve dar ordens ao computador.

## Programação Estruturada
Um programa é composto por blocos elementares de código que se interligam através de três mecanismos básicos: sequência, seleção e iteração.    
Cada uma destas construções tem um ponto de início e um ponto de término de execução.    
Aprogramação estruturada se tornou o paradigma mais usado para iniciar os estudos em programação, por conta da simplicidade de expressar os algoritmos.   
  
- **Sequência**: possui uma ordem de execução.
- **Seleção**: seleciona um trecho do código dentre diversas opções do fluxo.
- **Repetição**: iteração de diversos trechos de código.

**Função** - Retorna valor    
**Procedimento** - Apenas executa uma ação

```python
# Sequência
nome = input("Digite seu nome: ")
idade = int(input("Digite sua idade: "))
print("Olá,", nome, "você tem", idade, "anos.")
```

```python
# Seleção
idade = int(input("Digite sua idade: "))

if idade >= 18:
    print("Você é maior de idade.")
else:
    print("Você é menor de idade.")
```

```python
# Repetição (ou Estrutura de Loop) com for
for i in range(5):
    print(i)
```

```python
# Repetição (ou Estrutura de Loop) com While
contador = 0
while contador < 5:
    print(contador)
    contador += 1
```

```python
# Procedimentos
def saudacao():
    print("Olá, seja bem-vindo!")

# Chamando o procedimento
saudacao()
```

```python
# Função
def soma(a, b):
    return a + b

# Chamando a função
resultado = soma(3, 5)
print("O resultado da soma é:", resultado)
```

## Programação Orientada a Objetos

* Devido à sua fraca representatividade do mundo real, a programação estruturada foca na representação dos dados e operações desassociadas. Surgia a necessidade por programas que pudessem expressar melhor as características do dia-a-dia.     
* Na Programação Orientada a Objetos (POO), os programas são formados pela composição e interação entre diversas unidades chamadas de objetos.    
* Através dela é possível criar códigos mais próximos da forma como pensamos e agimos.



Um **objeto** pode ser considerado um encapsulamento de informações que representa algo do mundo real, sendo seu **estado** representado pelos valores armazenados em seus atributos. Além disso, os objetos são criados a partir das **classess**, que podem ser definidas como um template para objetos.



- **CLASSE**: Descreve um grupo de objetos com propriedades semelhantes, o mesmo comportamento, os mesmos relacionamentos com outros objetos e o mesmo significado no mundo real;
- **MÉTODO**: subprogramas (ou sub-rotinas) são encapsuladas nos próprios objetos e passam a designar-se método;
- **CONSTRUTOR**: É um recurso que serve para inicializar os atributos e é executado automaticamente sempre que um novo objeto é criado;
- **OBJETO**: É uma entidade que possui um estado e um conjunto definido de operações definidas para funcionar nesse estado.
- - **Estado de um objeto**: É um conjunto de atributos e respectivos valores de um objeto;
- **MULTIPLICIDADE**: Na associação entre classes, é o número de instâncias de uma classe que pode se relacionar com outra.
- **MÉTODO**: É a implementação de uma operação para uma classse.
- **MENSAGEM**: É a forma mais comum de ativar ou chamar os métodos. É quando um método chama ou usa o método de outro objeto.

##### Quatro pilares da POO



- **ABSTRAÇÃO**: Focalizar nos aspectos essenciais e inerentes a uma entidade e ignorar propriedades. Significa concenrar-se no que um objeto é e faz antes de decidir como ele será implementado.
- **ENCAPSULAMENTO**: É a caracteristica da OO capaz de ocultar partes (dados e detalhes) de implementação interna de classes do mundo exterior.
- **HERANÇA**: Uma subclasse X herda os atributos e métodos da superclasse Y.
  - A superclasse é uma **generalização** da subclasse
  - A subclasse é uma **especialização** da superclasse
  - **Herança multipla**: uma subclasse herda os atributos e métodos de mais de uma superclasse.
- **POLIMORFISMO**: É um princípio pelo qual duas ou mais classes derivadas de uma mesma classe podem invocar métodos que têm a mesma identificação (assinatura) mas comportamentos distintos.

```python
# Classe
class Carro:

    # Construtor
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        
    # Método
    def dirigir(self):
        print("O carro está sendo dirigido.")
```

```python
# Criando Objetos da classe Carro
meu_carro = Carro("Hyundai", "Hb20")
seu_carro = Carro("Toyota", "Corolla")
```

```python
print(meu_carro.modelo)
print(seu_carro.modelo)
```

```python
# Chamando o método dirigir
meu_carro.dirigir()  
```

```python
# Herança

# Superclasse
class Animal:
    def comer(self):
        print("O animal está comendo.")

# Subclasse
class Cachorro(Animal):
    def latir(self):
        print("O cachorro está latindo.")

# Criando um objeto da classe Cachorro
meu_cachorro = Cachorro()

meu_cachorro.comer()  # Métodos herdados da classe Animal
meu_cachorro.latir()  # Método da classe Cachorro
```

```python
# Polimorfismo

class Animal:
    def fazer_som(self):
        print("O animal faz um som.")

class Cachorro(Animal):
    def fazer_som(self):
        print("O cachorro late.")

class Gato(Animal):
    def fazer_som(self):
        print("O gato mia.")

# Função que aceita qualquer objeto Animal
def emitir_som(animal):
    animal.fazer_som()
```

```python
# Criando objetos das classes Cachorro e Gato
meu_cachorro = Cachorro()
meu_gato = Gato()

# Chamando a função emitir_som com diferentes objetos
emitir_som(meu_cachorro)  
emitir_som(meu_gato)     
```

```python
## Resposta 1:

# Solicita os dois números inteiros ao usuário
num1 = int(input("Digite o primeiro número inteiro: "))
num2 = int(input("Digite o segundo número inteiro: "))

# Calcula e imprime a soma, subtração, multiplicação e divisão
print("Soma:", num1 + num2)
print("Subtração:", num1 - num2)
print("Multiplicação:", num1 * num2)
print("Divisão:", num1 / num2)
```

```python
## Resposta 2:

class ContaBancaria:
    def __init__(self, titular, saldo=0):
        self.titular = titular
        self.saldo = saldo

    def depositar(self, valor):
        self.saldo += valor

    def sacar(self, valor):
        if self.saldo >= valor:
            self.saldo -= valor
        else:
            print("Saldo insuficiente.")

# Criando um objeto da classe ContaBancaria
minha_conta = ContaBancaria("João")

# Realizando operações de depósito e saque
minha_conta.depositar(1000)
minha_conta.sacar(500)

# Imprimindo o saldo resultante
print("Saldo atual:", minha_conta.saldo)
```
